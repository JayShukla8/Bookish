```python
import numpy as np
#we will now add 7 columns of all 7 emotions to each book and in each emotion of each book we will add the max score of that particular emotion in the description of that particular book
emotions = ["anger", "disgust", "fear", "joy", "neutral", "sadness", "surprise"]
isbns = []

#predictions is a list of lists containing the label-score for each sentence of the description of first book
#this function will return a dictionary of alphabetically arranged labels and the max score that emotion got in the description
def calculate_max_emotion_scores(predictions):
    per_emotion_score = {label: [] for label in emotions}
    for prediction in predictions:
        sorted_predictions = sorted(prediction, key=lambda x: x["label"])
        for index, label in enumerate(emotions):
            per_emotion_score[label].append(sorted_predictions[index]["score"])
    return {label: np.max(scores) for label, scores in per_emotion_score.items()}
```

```python
#now let us apply this function to each of the books
emotion_scores = {label: [] for label in emotions}
for i in range(10):
    isbns += [books["isbn10"][i]]
    sentences = books["description"][i].split('.')
    sentences = [item for item in sentences if item != ""]
    predictions = classifier(sentences)
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotions:
        emotion_scores[label].append(max_scores[label])
```

```python
#now that we know it works, let us apply it to all the books
from tqdm import tqdm
isbns = []
emotion_scores = {label: [] for label in emotions}

for i in tqdm(range(len(books))):
    isbns += [books["isbn10"][i]]
    sentences = books["description"][i].split('.')
    sentences = [item for item in sentences if item != ""]
    predictions = classifier(sentences)
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotions:
        emotion_scores[label].append(max_scores[label])
```

```python
emotion_scores_df = pd.DataFrame(emotion_scores)
emotion_scores_df["isbn10"] = isbns
emotion_scores_df
```

```python
books = pd.merge(books, emotion_scores_df, on="isbn10")
books
```
